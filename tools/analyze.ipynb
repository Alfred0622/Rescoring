# CLM

```python
with open(f"{result_dir}/tedlium2/noLM/dev/CLM_rerank_result.json") as f:
    data_list = json.load(f)
    hyps = []
    refs = []
    top_hyps = []
    totally_improved = 0
    partial_improved = 0
    partial_corrupt = 0
    totally_corrupt = 0
    neutral = 0
    same = 0
    errors = 0
    for data in data_list:
        hyps.append(data["rescore_hyps"])
        refs.append(data["ref"])
        top_hyps.append(data['top_hyps'])
    
        if (data["check_1"] == "Error"):
            errors += 1
        if (data["check_2"] == "Totally Improve"):
            totally_improved += 1
        elif (data["check_2"] == "Partial Improve"):
            partial_improved += 1
        elif (data["check_2"] == "Totally Corrupt"):
            totally_corrupt += 1
        elif (data["check_2"] == "Partial Corrupt"):
            partial_corrupt += 1
        elif (data['check_2'] == "Remain Correct"):
            same += 1
        elif (data['check_2'] == "Remain Error"):
            neutral += 1
    
    print(f'rescore hyp:{compute_measures(refs, hyps)}\n')
    print(f'top hyp:{compute_measures(refs, top_hyps)}')

    print(f"error:{errors}")
    print(f"totally_corrupt:{totally_corrupt}")
    print(f"partial_corrupt:{partial_corrupt}")
    print(f"totally_improved:{totally_improved}")
    print(f"partial_improved:{partial_improved}")
    print(f"same:{same}")
    print(f"neutral:{neutral}")

    print(f'bug check:{totally_corrupt + partial_corrupt + totally_improved + partial_improved + same + neutral}')
```
